 - Customer_ID,
  -  Product_ID,
   - Txn_Date,
   - Txn_Amount,
   - COUNT(*) AS duplicate_count

## Q8.Identify Sales_Transactions.Customer_ID values that violate referential integrity when joined with Customers_Master and write a query to detect such violations.

- SELECT DISTINCT s.Customer_ID

FROM Sale_Transaction s

LEFT JOIN Customers_Master c

ON s.Customer_ID = c.CustomerID

WHERE c.CustomerID IS NULL;
